```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
words = ['sum', 'summer', 'sumit', 'boller','bollyball','bolean'] 
```

```python
for i in words:
    print(i+' -- '+p_stemmer.stem(i)) #reduces each word to its root form.
```

Snowball Stemmmer

```python
from nltk.stem.snowball import SnowballStemmer

s_stemmer = SnowballStemmer(language='english') #15 language consider
```

```python
words = ['genration','genius','genuin']
```

```python
for i in words:
    print(i+ '--'+s_stemmer.stem(i))
```

```python
phrase = 'I am sumit'
for i in phrase.split():
    print(i+'--'+p_stemmer.stem(i))
```
